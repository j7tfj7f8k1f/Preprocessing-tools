```python
import os

def convert_to_utf8_with_lf(file_path):
    try:
        # 读取带有 CRLF (Windows 行尾符) 的 ANSI 编码文件
        with open(file_path, 'r', encoding='mbcs') as file:
            content = file.read()

        # 将 CRLF 替换为 LF
        content = content.replace('\r\n', '\n')

        # 以 UTF-8 编码并使用 LF 行尾符保存
        with open(file_path, 'w', encoding='utf-8', newline='\n') as file:
            file.write(content)

        print(f"转换成功: {file_path}")

    except Exception as e:
        print(f"转换失败: {file_path}, 错误: {e}")

def convert_folder(folder_path):
    # 遍历文件夹中的所有文件
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            if file_path.endswith('.txt'):  # 仅处理 .txt 文件，可以调整为其他类型
                convert_to_utf8_with_lf(file_path)

# 指定要转换的文件夹路径
folder_path = '/Users/User/Downloads/fb_files'
convert_folder(folder_path)
```

# 刪除沒有圖片跟文字檔案共存時

```python
import os
import shutil

def check_and_delete_folder(folder_path):
    has_image = False
    has_text = False

    # 遍歷資料夾中的所有文件
    for file in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file)
        if os.path.isfile(file_path):
            # 檢查是否有圖片文件 (.jpg, .png)
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                has_image = True
            # 檢查是否有文字文件 (.txt)
            if file.lower().endswith('.txt'):
                has_text = True

    # 如果沒有圖片和文字檔共存，則刪除整個資料夾
    if not (has_image and has_text):
        shutil.rmtree(folder_path)
        print(f"已刪除資料夾: {folder_path}")
    else:
        print(f"保留資料夾: {folder_path}")

def process_folders(base_folder):
    # 遍歷所有子資料夾
    for root, dirs, files in os.walk(base_folder):
        for dir_name in dirs:
            folder_path = os.path.join(root, dir_name)
            check_and_delete_folder(folder_path)

# 指定要處理的資料夾路徑
base_folder = '/content/drive/MyDrive'
process_folders(base_folder)
```

# 處理圖片大小


```python
from PIL import Image
import os

# 設定要轉換的圖像大小
new_size = (424, 424)

# 設定要轉換的根資料夾路徑
root_folder_path = '/content/drive/MyDrive'

# 遍歷根資料夾下的所有子資料夾
for subdir, dirs, files in os.walk(root_folder_path):
    for filename in files:
        # 如果不是圖像檔案，則跳過
        if not filename.endswith('.jpg') and not filename.endswith('.png'):
            continue

        # 設定圖片路徑
        image_path = os.path.join(subdir, filename)

        # 讀取圖片
        im = Image.open(image_path)

        # 計算圖片的長寬
        width, height = im.size

        # 計算縮放比例
        scale = min(424 / width, 424 / height)

        # 計算縮放後的圖片尺寸
        new_width = int(width * scale)
        new_height = int(height * scale)

        # 縮放圖片
        im_resized = im.resize((new_width, new_height), Image.BICUBIC)

        # 建立一個 640x640 的白色圖片
        im_white = Image.new('RGBA', (424, 424), (255, 255, 255, 255))

        # 將縮放後的圖片貼到白色圖片的中心
        im_white.paste(im_resized, (int((424 - new_width) / 2), int((424 - new_height) / 2)))

        # 覆蓋保存圖片
        im_white.save(image_path, 'PNG', quality=100, optimize=True)

print("圖片處理完成並覆蓋原本圖片！")
```

# gpu加速版本圖片大小

```python
import os
from PIL import Image
import torch
import torchvision.transforms as T

# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Define the image transformation pipeline
new_size = (640, 640)
transform = T.Compose([
    T.Resize(640),
    T.Pad((0, 0, 640, 640), padding_mode='constant', fill=(255, 255, 255, 255))  # pad to 640x640
])

# Path to the root folder containing images
root_folder_path = '/content/drive/MyDrive/chiikawa'

# Iterate through the root folder and process images
for subdir, dirs, files in os.walk(root_folder_path):
    for filename in files:
        # Skip non-image files
        if not filename.endswith('.jpg') and not filename.endswith('.png'):
            continue

        # Full path to the image
        image_path = os.path.join(subdir, filename)

        # Open image
        im = Image.open(image_path).convert('RGBA')

        # Convert image to PyTorch tensor and move to GPU
        im_tensor = T.ToTensor()(im).unsqueeze(0).to(device)

        # Apply transformations (resizing and padding) on GPU
        im_resized = transform(im_tensor)

        # Convert back to PIL image and move to CPU
        im_resized = T.ToPILImage()(im_resized.squeeze(0).cpu())

        # Save the resized image
        new_image_path = os.path.join(subdir, filename)
        im_resized.save(new_image_path, 'PNG', quality=100, optimize=True)

print("Image processing completed with GPU acceleration!")
```

# gemini api用來找出文字簡單說明

```python

import os
import time
import google.generativeai as genai
from google.colab import userdata

# 設置 Google API 金鑰
genai.configure(api_key=userdata.get("GOOGLE_API_KEY"))

# 設定根資料夾路徑
root_folder_path = '/content/drive/MyDrive'

# 初始化生成模型
model = genai.GenerativeModel("gemini-1.5-flash")

# 設定請求間隔時間（單位：秒）
request_interval = 15  # 例如，每 15 秒發送一個請求

# 遍歷根資料夾下的所有子資料夾及文字檔
for subdir, dirs, files in os.walk(root_folder_path, topdown=False):  # 使用topdown=False以便先處理文件再處理資料夾
    try:
        for filename in files:
            file_path = os.path.join(subdir, filename)

            # 檢查文件是否是文本檔案（假設所有文件都是文本文件）
            if filename.endswith('.txt'):
                try:
                    # 嘗試以 'utf-8' 打開，如果失敗則嘗試使用 'big5'
                    try:
                        with open(file_path, 'r', encoding='utf-8') as file:
                            file_content = file.read()
                    except UnicodeDecodeError:
                        with open(file_path, 'r', encoding='big5') as file:  # 使用 'big5' 重新打開
                            file_content = file.read()

                    # 使用生成模型進行內容生成
                    response = model.generate_content([f"請幫我找出這則貼文全部的商品名稱,請直接回應名稱就好", file_content])

                    # 獲取模型返回的文本
                    answer = response.text

                    # 生成新的文件名稱，並在名稱後加上 "_ok"
                    new_filename = filename.replace('.txt', '_ok.txt')
                    new_file_path = os.path.join(subdir, new_filename)

                    # 將生成的結果保存到新文件，使用 'utf-8' 編碼
                    with open(new_file_path, 'w', encoding='utf-8') as output_file:
                        output_file.write(answer)

                    # 刪除舊的文件
                    os.remove(file_path)

                    # 等待指定的時間間隔
                    time.sleep(request_interval)

                except Exception as e:
                    print(f"處理文件 {file_path} 時發生錯誤: {e}")
                    continue  # 出現錯誤後跳過當前文件，繼續處理下一個文件

        # 當子資料夾中的所有文件處理完成後，將資料夾名稱後加上 "_ok"
        parent_dir = os.path.dirname(subdir)
        new_subdir = subdir + "_ok"
        os.rename(subdir, new_subdir)

    except Exception as e:
        print(f"處理資料夾 {subdir} 時發生錯誤: {e}")
        continue  # 出現錯誤後跳過當前資料夾，繼續處理下一個

print("處理完成，文件及資料夾名稱已修改並保存！")
```

# gpt處理完後要刪除特殊符號

```python
import os
import re

# 設定根資料夾路徑
root_folder_path = '/content/drive/MyDrive'

# 正則表達式：保留中文字、英文字母、數字、空格、逗號（,）和頓號（、）
pattern = re.compile(r'[^\u4e00-\u9fffA-Za-z0-9\s,、]')

# 遍歷根資料夾下的所有子資料夾及文字檔
for subdir, dirs, files in os.walk(root_folder_path):
    for filename in files:
        # 只處理文字檔案（假設是 .txt 檔案）
        if filename.endswith('.txt'):
            file_path = os.path.join(subdir, filename)

            # 讀取原文字檔內容
            with open(file_path, 'r', encoding='utf-8') as f:
                content = f.read()

            # 只保留數字、英文字母、中文字、空格、逗號和頓號，移除其他符號
            cleaned_content = pattern.sub('', content)

            # 將處理過的內容覆蓋回原文字檔
            with open(file_path, 'w', encoding='utf-8') as f:
                f.write(cleaned_content)

print("文字檔案處理完成並覆蓋原本文字檔！")
```

# 做出單詞表


```python
import os

# 指定資料夾路徑
folder_path = '/content/drive/MyDrive'
output_path = '/content/word_list.txt'

# 儲存所有單詞
word_set = set()

# 遍歷資料夾中的所有檔案
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.txt'):  # 檢查檔案是否為文字檔
            file_path = os.path.join(root, file)
            with open(file_path, 'r', encoding='utf-8') as f:
                # 讀取檔案內容
                content = f.read()
                # 將內容分割成單詞，並加入集合中
                words = content.split()
                word_set.update(words)

# 將單詞寫入檔案
with open(output_path, 'w', encoding='utf-8') as f:
    for word in sorted(word_set):  # 排序單詞
        f.write(f"{word}\n")

print(f"單詞表已儲存至 {output_path}")
```

# 重新命名讓他乾淨點

```python
import os
import re

def rename_folders(root_folder):
    for folder_name in os.listdir(root_folder):
        folder_path = os.path.join(root_folder, folder_name)

        # 只處理資料夾
        if os.path.isdir(folder_path):
            # 移除名稱中的 "ok" 並只保留中文和英文
            new_folder_name = re.sub(r"ok", "", folder_name, flags=re.IGNORECASE)
            new_folder_name = re.sub(r"[^a-zA-Z\u4e00-\u9fff]+", "", new_folder_name)

            # 如果名稱有改變才重新命名
            if new_folder_name != folder_name:
                new_folder_path = os.path.join(root_folder, new_folder_name)
                os.rename(folder_path, new_folder_path)
                print(f"已將 '{folder_name}' 重新命名為 '{new_folder_name}'")

# 設定要處理的根資料夾路徑
root_folder_path = '/content/drive/MyDrive'

# 執行重新命名函數
rename_folders(root_folder_path)

print("資料夾名稱處理完成！")
```
